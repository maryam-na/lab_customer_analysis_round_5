```python
# Standardization of header names
def standardization(df):
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ','_')
```

```python
standardization(data)
data
```

```python
#Check and deal with NaN values.
def nulls_data(df):
    nulls_df = pd.DataFrame(round(df.isna().sum()/len(df)*100,2))
    nulls_df = nulls_df.reset_index()
    nulls_df.columns = ['header_name', '%_null_values'] 
    return nulls_df
```

```python
res_null=nulls_data(data)
res_null
```

```python
columns_drop = res_null[res_null['%_null_values']>5]['header_name']
columns_drop
```

```python
# The column 'vehicle_type' is dropped because of the high percentage of missing data
data = data.drop(['vehicle_type'],axis=1)
# The numerical columns with the percentage of missing data higher than 5 are dropped
data = data.drop(['months_since_last_claim', 'number_of_open_complaints'],axis=1)
```

```python
data
```

```python
# Fixing NAN values which are categorical and replacing them with 'None'
def null_cat_fix(df):
    cat_cols= df.select_dtypes(include=['object']).columns.tolist()
    for col in cat_cols:
        df[col] = df[col].fillna('None')
    return df    
    
```

```python
#For the numerical variables, check the multicollinearity between the features. Please note that we will use the column total_claim_amount later as the target variable.
# correlation between different variables
f, ax = plt.subplots(figsize=(20, 15))
mat = numerical.corr('pearson')
mask = np.triu(np.ones_like(mat, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(mat, mask=mask, cmap=cmap, vmax=1, center=0, annot = True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
# Another way to check -VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
X_variables = numerical.drop(['total_claim_amount'],axis=1)
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X_variables.columns
vif_data["VIF"] = [variance_inflation_factor(X_variables.values, i) for i in range(len(X_variables.columns))]
vif_data
```

```python
# These VIF values should be checked to be less than max of 10 and 1/1-R²model
```

```python
# From features three columns have outliers which should be corrected
# column customer_lifetime_value-removing outliers
x['customer_lifetime_value'].unique()
```

```python
iqr=np.percentile(x['customer_lifetime_value'],75)-np.percentile(x['customer_lifetime_value'],25)
upper_limit=np.percentile(x['customer_lifetime_value'],75) +1.5*iqr
lower_limit=np.percentile(x['customer_lifetime_value'],75) -1.5*iqr
upper_limit, lower_limit
```

```python
x = x[(x['customer_lifetime_value']>lower_limit)&(x['customer_lifetime_value']<upper_limit)]
```

```python
cols=['customer_lifetime_value',
 'income',
 'monthly_premium_auto',
 'months_since_policy_inception',
 'number_of_policies']
for col in cols:
   # x[col] = x[x[col] > 0]
    x[col] =np.where(x[col]<=0, np.mean(x[col]), x[col]) 
```

```python
x
```

```python
# Normalization of numerical data usig log transform
def log_transform_clean(input):
    log_input = np.log(input)
    if np.isfinite(log_input):
        return log_input
    else:
        return np.mean(log_input)
```

```python
x_log= x.applymap(log_transform_clean)   
```

```python
x_log
```

```python
# checking the distribution after log_transformation
for col in cols:
    sns.distplot(x_log[col],bins= 100, 
                   color = 'blue')
    plt.title('Distribution')
    plt.show()
```

```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(x)
x_normalized = transformer.transform(x)
print(x_normalized)
```
